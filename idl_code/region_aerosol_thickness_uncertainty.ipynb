```python
#=== this script is to derive regional aerosol layer thickness and opticl depth uncertainty 
#=== the layer thickess is segrated into two optical depth area tau > 0.2 and tau < 0.2
#=== test two month data, July and Jan
import numpy as np
import matplotlib.pyplot as plt
from pyhdf.SD import SD, SDC
import glob
from datetime import datetime, date, timedelta

print('start',datetime.now())

datadir='/data/gdi/f/yulanh/CALIPSO/MLay/2018/'

fnames=glob.glob(datadir+'*2018-08*')
Nf=len(fnames)
print(Nf)
#define the region
#region='India'
#minlat=10
#maxlat=25
#minlon=70
#maxlon=85
#region='indonesia'
#minlat=-10
#maxlat=10
#minlon=95
#maxlon=150

#region='Arab'
#minlat=20
#maxlat=35
#minlon=45
#maxlon=65

region='WUS'
minlat=30
maxlat=45
minlon=-140
maxlon=-123

fnames=fnames

wfile=open('Sep_wus1808_aerosol_layerthickness_tau_tauerr.txt','w')

for fname in fnames:
    #print(fname)
    # read lidar data
    hdf=SD(fname)
    lat=hdf.select('Latitude')
    lat=lat[:,:]
    lon=hdf.select('Longitude')
    lon=lon[:,:]
    lidtop=hdf.select('Layer_Top_Altitude')
    lidtop=lidtop[:,:]
    lidbase=hdf.select('Layer_Base_Altitude')
    lidbase=lidbase[:,:]
    cad=hdf.select('CAD_Score')
    cad=cad[:,:]
    tau=hdf.select('Column_Optical_Depth_Tropospheric_Aerosols_532')
    tau=tau[:]
    tauerr=hdf.select('Column_Optical_Depth_Tropospheric_Aerosols_Uncertainty_532')
    tauerr=tauerr[:]
    Naero_cldlayer=hdf.select('Number_Layers_Found')
    Naero_cldlayer=Naero_cldlayer[:]
    
    lat=lat[:,1]
    lon=lon[:,1]
    for si in range(len(lat)):
        if ((lat[si] >= minlat) & (lat[si] <= maxlat) & (lon[si] >= minlon) & \
            (lon[si] <= maxlon) & (tau[si] > 0)):
            tpcad    =cad[si,:]
            tplidtop =lidtop[si,:]
            tplidbase=lidbase[si,:]
            tptau    =tau[si]
            tptauerr =tauerr[si]
            layer_ind=np.where(tplidtop > 0)[0]
            
            layerthick=0.0
            cldflag=0
            filter_flag=0
               
            cadfilter=np.where((tpcad >=-20) & (tpcad <0))[0]
            if len(cadfilter) > 0:
                filter_flag=1
            
            for hi in range(len(layer_ind)):
                if (tpcad[hi] >= -100) & (tpcad[hi] <= -20):
                    layerthick=layerthick+(tplidtop[hi]-tplidbase[hi])
                    
            cldind = np.where((tpcad >= 20) & (tpcad <=100))[0]
            if (len(cldind) > 0):
                cldflag=1
            
            if ((filter_flag == 0)):
                 wfile.write(str(cldflag)+' '+str(tptau[0])+' '+str(tptauerr[0])+' '+str(layerthick)+' '+fname+'\n')
            #print(si,tpcad)
            #print(cldflag,filter_flag,layerthick,tptau,tptauerr)
wfile.close()            
print('finish',datetime.now())
```
